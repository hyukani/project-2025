```python
from fredapi import Fred
import pandas as pd

# API (di https://fred.stlouisfed.org/)
fred = Fred(api_key="API KEY PRIBADI")

series = [
    "UNRATE","AHETPI","PERMIT","CSCICP03USM665S","AAA10Y",
    "M2REAL", "CPIAUCSL", "FEDFUNDS", "PAYEMS", "PCE",
    "GS10", "T10Y2Y", "DGS10", "CPILFESL", "HOUST",
    "WPSFD49207", "GS1", "DSPIC96", "INDPRO", "CPIAUCSL", # Ubah kode "SPCPI" dengan "CPIAUCSL"
    "IC4WSA", "IPMAN", "WM2NS"
]

df_list = []
for s in series:
    try:
        df = fred.get_series(s, observation_start='1984-01-01', observation_end='2024-12-31')
        df.name = s
        df_list.append(df)
    except ValueError as e:
        print(f"Error fetching data for series '{s}': {e}")

data = pd.concat(df_list, axis=1).reset_index().rename(columns={"index":"DATE"})

usrec = fred.get_series("USREC", observation_start='1984-01-01', observation_end='2024-12-31')
usrec = usrec.reset_index().rename(columns={"index":"DATE","USREC":"RECESSION"})

df_final = pd.merge(data, usrec, on="DATE")

# Simpan CSV
df_final.to_csv("recession_dataset_1984_2024.csv", index=False)
print("Saved to recession_dataset_1984_2024.csv")
```
